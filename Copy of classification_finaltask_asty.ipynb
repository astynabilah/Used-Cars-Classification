##Convert data into dataframe

```
import pandas as pd

df_nb = pd.read_csv('used_cars_processed.csv',engine="python")
pd.set_option('display.max_columns', None)

df_nb= df_nb.loc[:, df_nb.columns != 'Unnamed: 0'] #dropping index column
df_nb
```

##Making 'price' feature into categorical
Price divided by two categories: <= mean and < mean

If below and equals to mean: 0
If above mean: 1

```
mean_price = df_nb.price.mean()

df_nb.loc[df_nb.price <= mean_price, 'price'] = 0
df_nb.loc[df_nb.price > mean_price, 'price'] = 1

# df_nb
```

##Splitting data into training and testing datasets

```
from sklearn.model_selection import train_test_split

df_nb2 = df_nb.copy()
y_nb = df_nb2.pop('price').values
df_nb_train, df_nb_test = train_test_split(df_nb, train_size= 0.7,test_size= 0.3)
y_nb_train = df_nb_train.pop('price').values
y_nb_test = df_nb_test.pop('price').values
df_nb_train, df_nb_test, y_nb_train, y_nb_test
```

##Model: Using GaussianNB

```
from sklearn.naive_bayes import GaussianNB
from imblearn.combine import SMOTEENN

nb_model = GaussianNB()
smote_enn = SMOTEENN(random_state=100)
df_nb_train, y_nb_train = smote_enn.fit_resample(df_nb_train, y_nb_train)
nb_model.fit(df_nb_train, y_nb_train)
y_nb_pred = nb_model.predict(df_nb_test)
```

##Result of Prediction

```
res = df_nb_test.copy()
res['actual class'] = y_nb_test
res['predicted class'] = y_nb_pred
res
```

##Evaluation: Accuracy -- Cross Validation

```
from sklearn.model_selection import cross_val_score

scores = cross_val_score(nb_model, df_nb2, y_nb, cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

# KNN Classification

## Converting data into dataframe

```
import pandas as pd

df_knn = pd.read_csv('used_cars_processed.csv',engine="python")
pd.set_option('display.max_columns', None)

df_knn= df_knn.loc[:, df_knn.columns != 'Unnamed: 0'] #dropping index column
# df_knn
```

##Making 'price' feature into categorical
Price divided by two categories: <= mean and < mean

If below and equals to mean: 0
If above mean: 1

```
mean_price = df_knn.price.mean()

df_knn.loc[df_knn.price <= mean_price, 'price'] = 0
df_knn.loc[df_knn.price > mean_price, 'price'] = 1
df_knn
```

##Splitting data into training and testing datasets

```
from sklearn.model_selection import train_test_split
df_knn2 = df_knn.copy()
y_knn = df_knn2.pop('price').values
df_knn_train, df_knn_test = train_test_split(df_knn, train_size= 0.7,test_size= 0.3)
y_knn_train = df_knn_train.pop('price').values
y_knn_test = df_knn_test.pop('price').values

df_knn_train, df_knn_test, y_knn_train, y_knn_test
```

##Model: Using KNeighborsClassifier

```
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=2)
knn_model.fit(df_knn_train, y_knn_train)
y_knn_pred = knn_model.predict(df_knn_test)
```

##Result of Prediction

```
res = df_knn_test.copy()
res['actual class'] = y_knn_test
res['predicted class'] = y_knn_pred
res
```

##Evaluation: using cross validation and accuracy to make it more precise

```
from sklearn.model_selection import cross_val_score

scores = cross_val_score(knn_model, df_knn2, y_knn, cv=10)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
